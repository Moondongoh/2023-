```python
import pandas as pd
import numpy as np

# '일최저체감온도(°C)', '일최고기온(°C)', '일평균기온(°C)'에 대한 중간값 계산
crosstab = pd.crosstab(df['일최저체감온도(°C)'], df['일최고기온(°C)'], values=df['일평균기온(°C)'], aggfunc=np.median)

# 결과 확인
print(crosstab)
```

```python
#pick한 열 정보
selected_columns = ['일시', '일최고기온(°C)', '일최저기온(°C)', '일최저체감온도(°C)']
df_selected = df[selected_columns]

#일시 열을 일시 형식으로 변환
df_selected['일시'] = pd.to_datetime(df_selected['일시'])

#평균을 계산해줌
grouped_data = df_selected.groupby('일시').mean()

#그래프 그리기
grouped_data.plot(kind='line', figsize=(12, 6))
plt.title("일별 평균 온도")
plt.xlabel("일시")
plt.ylabel("평균 온도 (°C)")
plt.legend(["일최고기온", "일최저기온", "일최저체감온도"])
plt.grid(True)
plt.show()
```
